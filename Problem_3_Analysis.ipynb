```python
import pandas as pd
import lxml.etree,xml.etree.ElementTree as et
```

```python
%cd /content/drive/MyDrive/Vehicle_Detection/Kaggle/train/Final Train Dataset
```

```python
import glob,os
xml_files = []
# for file in glob.glob("Data/train/Img/*xml"):
for file in glob.glob("*xml"):
    xml_files.append(os.path.basename(file))
```

```python
xml_files.remove('231.xml')
```

```python
# XML to pandas dataframe conversion
a = {'folder': [],
     'filename': [],
     'path': [],
     'database': [],
     'width': [],
     'height': [],
     'depth': [],
     'segmented': [],
     'name': [],
     'pose': [],
     'truncated': [],
     'difficult': [],
     'xmin': [],
     'ymin': [],
     'xmax': [],
     'ymax': []}


for file_name in xml_files:
    etree = et.parse(file_name)
    eroot = etree.getroot()

    doc = lxml.etree.parse(file_name)
    count = doc.xpath('count(//object)')

    for x in eroot.iter():
        if x.tag in a.keys():
            a[x.tag].append(x.text)
        if (x.tag in ['folder','filename','path','database','width','height','depth','segmented']):
            for i in range (1,int(count)):
              a[x.tag].append(x.text)
```

```python
df = pd.DataFrame.from_dict(a)
df
```
